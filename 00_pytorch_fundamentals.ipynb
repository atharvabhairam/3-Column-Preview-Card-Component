#Reshaping, stacking, squeezing, unsqueezing tensors
*Reshaping - reshapes an input tensor to a defined shape

*View- Return a view of an input tensor of certain shape but keep the same memory as the original tensor

*Stacking- combine multiple tensors on top of each other(vstack) or side by side (hstack)

*Squeeze- removes all `1` dimensions from a tensor

*Unsqueeze- add a `1` dimension to a target tensor

*Permute- Return a view of the input with dimensions permuted (swapped) in a certain way.

```python
#Let's create a tensor
import torch
x=torch.arange(1., 10.)
x, x.shape
```

```python
#Add an extra dimension
x_reshaped=x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

```python
#Change the view
z=x.view(1,9)
z, z.shape
```

```python
z[:,0]=5
z, x
```

```python
x_stacked=torch.stack([x, x, x, x],dim=0)
x_stacked
```

```python
#torch.squeeze()- removes all single dimensions from a target tensor
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

#Remove extra dimensions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python

x_reshaped.shape
```

```python
x_reshaped.squeeze()
```

```python
x_reshaped.squeeze().shape
```

```python
# torch.unsqueeze() - adds a single dimension to get a target tensor at a specfic dim(dimension)

print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed}")

#Remove extra dimensions from x_reshaped
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
#torch.permute - rearranges the dimensions of a target tensor in a specified order
x_original=torch.rand(size=(224,224,3)) #[height, width, colour_channels]

#Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2,0,1) #shifts axis 0->1,1->2,2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") #[colour_channels, height, width]
```

## PyTorch tensors & Numpy
Numpy is a popular scientific Python Numerical computing library.
And because of this, PyTorch has functionality to interact with it.

*Data is Numpy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`

*Pytorch tensor-> Numpy -> `tensor.numpy()`

```python
# NumPy ↔ PyTorch Memory Sharing (Full Explanation + Code)


import numpy as np
import torch

print("=== 1) NumPy -> PyTorch (torch.from_numpy) ===")

# Create a NumPy array
array = np.arange(1.0, 8.0)

# Convert NumPy array to PyTorch tensor
tensor = torch.from_numpy(array)

print("Original array:", array)
print("Tensor from numpy:", tensor)

# -------------------------------------------------------
# NOTE:
# torch.from_numpy() shares the SAME memory.
# Both point to the same data UNTIL YOU CREATE A NEW ARRAY.
# -------------------------------------------------------

print("\n--- Modify NumPy array OUT-OF-PLACE (array = array + 1) ---")

# OUT-OF-PLACE OPERATION → creates NEW array → breaks connection
array = array + 1

print("Updated array:", array)
print("Tensor remains SAME:", tensor)

# -------------------------------------------------------
# NOTE:
# array = array + 1 creates a NEW array → tensor DOES NOT update.
# -------------------------------------------------------

print("\n--- Modify NumPy array IN-PLACE (array2 += 1) ---")

array2 = np.arange(1.0, 8.0)
tensor2 = torch.from_numpy(array2)

array2 += 1   # IN-PLACE update → shared memory maintained

print("array2 after in-place update:", array2)
print("tensor2 also updates:", tensor2)

# -------------------------------------------------------
# NOTE:
# array2 += 1 updates existing memory → tensor2 ALSO updates.
# -------------------------------------------------------

print("\n=== 2) PyTorch -> NumPy (tensor.numpy()) ===")

# Create a tensor
tensor3 = torch.ones(7)

# Convert to NumPy
numpy_tensor3 = tensor3.numpy()

print("tensor3:", tensor3)
print("numpy_tensor3:", numpy_tensor3)

# -------------------------------------------------------
# NOTE:
# tensor.numpy() shares the SAME memory.
# -------------------------------------------------------

print("\n--- Modify Tensor OUT-OF-PLACE (tensor3 = tensor3 + 1) ---")

tensor3 = tensor3 + 1   # Creates NEW tensor → breaks connection

print("tensor3 updated:", tensor3)
print("numpy_tensor3 unchanged:", numpy_tensor3)

# -------------------------------------------------------
# NOTE:
# tensor3 = tensor3 + 1 creates NEW memory → numpy_tensor3 does NOT update.
# -------------------------------------------------------

print("\n--- Modify Tensor IN-PLACE (tensor4.add_(1)) ---")

tensor4 = torch.ones(7)
numpy_tensor4 = tensor4.numpy()

tensor4.add_(1)   # IN-PLACE update

print("tensor4 after in-place:", tensor4)
print("numpy_tensor4 also updated:", numpy_tensor4)

# -------------------------------------------------------
# NOTE:
# tensor4.add_(1) updates original memory → numpy_tensor4 ALSO updates.
# -------------------------------------------------------

print("\n=== SUMMARY ===")
print("In-place ops keep memory shared (+= , add_()).")
print("Out-of-place ops break the sharing (array + 1, tensor + 1).")
```
